### **Define functions for calculating the different features**

```
import pandas as pd
```

```

import pandas as pd

from Pfeature.pfeature import aac_wp, dpc_wp

# Amino acid composition (AAC)

def aac(input):
    a = input.rstrip('txt')
    output = a + 'aac.csv'
    df_out = aac_wp(input, output)
    df_in = pd.read_csv(output)
    return df_in

# Dipeptide composition (DPC)

def dpc(input):
    a = input.rstrip('txt')
    output = a + 'dpc.csv'
    df_out = dpc_wp(input, output, 1)
    df_in = pd.read_csv(output)
    return df_in



```

### **Calculate feature for both positive and negative classes + combines the two classes + merge with class labels**

```
pos = 'train_po_cdhit.txt'
neg = 'train_ne_cdhit.txt'

def feature_calc(po, ne, feature_funcs):
    po_features = []
    ne_features = []
    for feature_func in feature_funcs:
        po_feature = feature_func(po)
        ne_feature = feature_func(ne)
        po_features.append(po_feature)
        ne_features.append(ne_feature)
    po_combined = pd.concat(po_features, axis=1)
    ne_combined = pd.concat(ne_features, axis=1)
    po_class = pd.Series(['positive' for i in range(len(po_combined))])
    ne_class = pd.Series(['negative' for i in range(len(ne_combined))])
    po_ne_class = pd.concat([po_class, ne_class], axis=0)
    po_ne_class.name = 'class'
    po_ne_feature = pd.concat([po_combined, ne_combined], axis=0)
    df = pd.concat([po_ne_feature, po_ne_class], axis=1)
    return df

# Specify the features to be calculated
features_to_extract = [aac, dpc]

# Calculate and combine features
feature = feature_calc(pos, neg, features_to_extract)
```

```
# Separate features and labels
X = feature.drop('class', axis=1)
y = feature['class'].copy()


X.shape

# Encode the labels
y = y.map({"positive": 1, "negative": 0})


# Feature selection (Variance threshold)
from sklearn.feature_selection import VarianceThreshold

fs = VarianceThreshold(threshold=0.1)
fs.fit_transform(X)
X2 = X.loc[:, fs.get_support()]
X2
```
